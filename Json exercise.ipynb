```python
#Excercise 2: Find the top 10 major project themes (using column 'mjtheme_namecode')

#load json file as string
import json
data=json.load(open('C:/Users/Aditya/Downloads/json project/data_wrangling_json/data/world_bank_projects.json'))
project_themes = json_normalize(data,record_path='mjtheme_namecode')


project_themes
```

```python
#excercise 3:In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

#Sort Values in the dataframe
project_themes.sort_values('code', inplace= True)
```

```python
# Need to match codes to names. In order to do so we can create unique lists of both the columns and zip them to a dictionary
a = project_themes.code.unique().tolist()
filter1 = (project_themes.name != "")
b = project_themes[filter1].name.unique().tolist()
assignment = dict(zip(a, b))
```

```python
#First replace white spaces with NaN and then fill those spaces with the correct assignment
project_themes.name = project_themes.name.replace('', np.nan)
project_themes['name'] = project_themes['name'].fillna(project_themes['code'].map(assignment))
```

```python
#Checking for null values
project_themes['name'].isna().value_counts()
```
